## Prepare environment

```python
import gymnasium as gym
gym.register(
    id="FrozenLakeEasy-v0",
    entry_point="gymnasium.envs.toy_text:FrozenLakeEnv",
    kwargs={"is_slippery": False}
)
```

```python
from datetime import datetime
from gymnasium.wrappers import RecordVideo

env = RecordVideo(
    gym.make("FrozenLakeEasy-v0", render_mode = "rgb_array"),
    "./recorded/{}".format(datetime.now().strftime('%Y%m%d%H%M')),
    disable_logger=True
)
```

## Train agent

```python
agent = mc.MonteCarloAgent(epsilon=0.01)
agent.learn(env)

FlowLayout().draw([
    plot_q_value(agent.Q),
    agent.plot_reward_log()
])
```

```python
agent = mc.MonteCarloAgent(epsilon=0.1)
agent.learn(env)

FlowLayout().draw([
    plot_q_value(agent.Q),
    agent.plot_reward_log()
])
```
